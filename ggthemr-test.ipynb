```r
# System information
sessionInfo()

# Import libraries
library(tidyverse)
library(ggthemr)

# Plotting Options
options(repr.plot.width=6, repr.plot.height=4)
options(repr.plot.res=300, repr.plot.pointsize=12)
options(jupyter.plot_mimetypes = "image/png")
```

```r
# Following the example from:
# https://github.com/cttobin/ggthemr#a-note-about-theme-setting
ggthemr_reset()
dust_theme <- ggthemr('dust', set_theme = FALSE)
```

```r
# Default Plot
# Observation: As expected, no change in default theme.
ggplot(data = mpg) + 
geom_point(mapping = aes(x = displ, y = hwy, color = class), size=2)
```

```r
# Apply Theme
# Observation: As expected, Dust theme is applied. Consistent with instructions.
ggplot(data = mpg) + 
    geom_point(mapping = aes(x = displ, y = hwy, color = class), size=2) + 
    dust_theme$theme
```

```r
# Apply Theme + Scale Colors
# Observation: Dust scale colors not applied. Inconsistent with the manual.
ggplot(data = mpg) + 
    geom_point(mapping = aes(x = displ, y = hwy, color = class), size=2) + 
    dust_theme$theme + 
    dust_theme$scales$scale_fill_discrete()
```

```r
# Apply Scale Colors
# Observation: Expected behavior to just apply the scale colors, but what we get is default plot.
ggplot(data = mpg) + 
    geom_point(mapping = aes(x = displ, y = hwy, color = class), size=2) + 
    dust_theme$scales$scale_fill_discrete()
```
